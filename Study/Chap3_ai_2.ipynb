- 옵션 : 기초자산을 미리(현재) 정한 가격으로 미래 일정시점에 가서 사거나 팔 수있는 권리(가 내재된 파생상품), 권리의 대가로 프리미엄(옵션가격) 지불<br>

1. 현물 상승 예상 
- 내가 싸게 살 수 있는 권리를 행사 -> call 매수
- 상대방의 비싸게 팔 권리가 소용없게 만드는 -> put 매도

2. 현물 하락 에상
- 내가 비싸게 팔 수 있는 권리를 행사 -> put 매수
- 상대방에게 싸게 살 권리가 소용없게 -> call 매도 

- 옵션의 가치(시간 가치) = 내재가치 + 시간가치
- 시간가치 : 만기까지의 기간에서 기초자산의 가격 변화에 따라 이익을 얻을 수 있는 가능성에 대한 가치 -> 미래의 주가 변동에 따라 기대되는 프리미엄
- 내재가치 : 지금 권리를 행사 할 때의 옵션 가치(현재의 기초자산 가격과 행사의 가격차이로 측정), 옵션 본연의 가치 내가격 상태에서만 존재


옵션의 4가지 기능 :
1. 헷지 : 만기일에 기초자산의 가격변동에서 오는 위험을 회피할 수 있는 기능
2. 투자위험 제한 : 옵션을 매수하면 이익의 폭은 제한이 없으나 손실은 프리미엄으로 제한(보험효과)
3. 다양한 투자전략 : 기초자산의 가격변동과 연계하여 콜/풋, 만기일, 행사가격 등을 선택, 다양한 투자전략 구사
4. 레버리지 효과 : 적은 투자자금으로 큰 이익을 얻을 수 있는 레버리지가 큰 투자수단

- 내가격 (Bid Price):

매수 주문을 하는 투자자나 거래소가 주식이나 증권을 사려는 가격입니다.<br>
내가격은 투자자가 얼마의 가격으로 주식을 사고자 하는지를 나타냅니다.<br>
내가격은 보통 거래소에 표시되어 있고, 시장가격과는 다를 수 있습니다.<br>



- 외가격 (Ask Price 또는 Offer Price):

매도 주문을 하는 투자자나 거래소가 주식이나 증권을 팔려는 가격입니다.<br>
외가격은 투자자가 얼마의 가격으로 주식을 팔고자 하는지를 나타냅니다.<br>
외가격 또한 거래소에 표시되어 있고, 시장가격과는 다를 수 있습니다.<br>

<br>간단히 말하면, 내가격은 주식을 사고자 하는 투자자가 제시하는 가격이고, 외가격은 주식을 팔고자 하는 투자자가 제시하는 가격입니다.<br> 이 두 가격의 차이는 스프레드(Spread)라고도 불리며, 이는 거래의 유동성과 거래 비용을 나타내는 중요한 지표 중 하나입니다.<br> 스프레드가 작을수록 시장의 유동성이 높아진다고 할 수 있습니다.

예를 들어, 주식 A의 내가격이 100원이고 외가격이 102원이라면, 이는 투자자들이 주식 A를 100원에 사고자 하는데 반해, 판매자들은 102원에 팔고자 한다는 것을 의미합니다. 따라서 내가격과 외가격 사이의 차이인 스프레드는 2원입니다.

행사가격과 기초자산 가격과의 대소관계
- 등가격옵션(at-the-money ATM) :기초자산가격 = 행사가격
- 내자격옵션(in-the-money ITM) : 당장행사 -> 양의 payoff (매수자가 원하는 가격)
1. 콜 : 기초자산가격 > 행사가격
2. 풋 : 기초자산가격 < 행사가격
- 외가격옵션(out-the-money OTM) : (매도자가 원하는 가격)
1. 콜 : 기초자산가격 < 행사가격
2. 풋 : 기초자산가격 > 행사가격

옵션 계약에는 주로 두 가지 유형이 있습니다: 콜옵션(Call Option)과 풋옵션(Put Option). 또한, 옵션의 행사 방식에 따라 유로피안 옵션(European Option)과 아메리칸 옵션(American Option)으로 나뉩니다. 이들을 간단히 설명하겠습니다.

1. 콜옵션 (Call Option):

- 콜옵션은 옵션 계약의 한 유형으로, 특정 자산(주식, 환율, 상품 등)을 일정 가격(행사 가격)으로 일정 기간 동안 매입할 수 있는 권리를 제공합니다.
- 콜옵션 소유자는 행사일에 자산을 매입할지 여부를 결정할 수 있습니다. 자산의 가격이 행사 가격보다 높은 경우에만 유리하므로, 주로 가격 상승을 예상할 때 매수됩니다.
- 콜옵션은 주식 시장에서는 주식을 매수하는 권리를, 기타 시장에서는 해당 자산을 매수하는 권리를 제공합니다.
<br> 

2. 풋옵션 (Put Option):

- 풋옵션은 옵션 계약의 다른 유형으로, 특정 자산을 일정 가격으로 일정 기간 동안 매도할 수 있는 권리를 제공합니다.
- 풋옵션 소유자는 행사일에 자산을 매도할지 여부를 결정할 수 있습니다. 자산의 가격이 행사 가격보다 낮은 경우에만 유리하므로, 주로 가격 하락을 예상할 때 매수됩니다.
- 풋옵션은 주식 시장에서는 주식을 매도하는 권리를, 기타 시장에서는 해당 자산을 매도하는 권리를 제공합니다.

3. 유로피안 옵션 (European Option):

- 유로피안 옵션은 행사 권리를 특정 행사일에만 행사할 수 있는 옵션입니다. 따라서 행사일에만 옵션을 행사할 수 있습니다.
- 유로피안 콜옵션은 주식을 일정 가격으로 매입하는 권리를, 풋옵션은 주식을 일정 가격으로 매도하는 권리를 행사할 수 있습니다.

4. 아메리칸 옵션 (American Option):

- 아메리칸 옵션은 행사 권리를 행사일 전에 언제든지 행사할 수 있는 옵션입니다. 따라서 유로피안 옵션보다 유연성이 높습니다.
- 아메리칸 콜옵션은 주식을 일정 가격으로 매입하는 권리를, 아메리칸 풋옵션은 주식을 일정 가격으로 매도하는 권리를 언제든지 행사할 수 있습니다.

- 알파, 베타란?
1. 알파 -> 알파는 주식 투자의 초과 수익률을 나타내는 지표입니다. 알파는 기대 수익률과 기준 지수(일반적으로 주식 시장 전반을 나타내는 지수)의 예상 수익률 간의 차이를 나타냅니다. 양수인 알파는 투자자가 시장 평균보다 초과 수익을 올린 것을 의미하며, 음수인 알파는 시장 평균보다 수익이 뒤쳐진 것을 나타냅니다.
2. 베타 -> 베타는 투자자의 포트폴리오와 시장 전체의 움직임 사이의 관계를 나타내는 지표입니다. 베타는 주식 또는 포트폴리오의 변동성이 시장 전반적인 변동성과 얼마나 연관되어 있는지를 측정합니다. 베타가 1에 가까우면 시장과의 상관관계가 높다는 의미이며, 1보다 작으면 시장보다 변동성이 낮은 것을 의미하고, 1보다 크면 시장보다 변동성이 높은 것을 나타냅니다.

- 무차익 거래란? -> <br>
무차익 거래(arbitrage)는 금융 시장에서 동일한 자산이나 계약을 다른 시장에서 저렴하게 매입하여 더 비싼 시장에서 판매하여 이익을 창출하는 전략. 
1. 가격 차이 발견 : 두 개 이상의 시장에서 동일한 자산의 가격차이 발견
2. 매매실행 : 저렴한 시장에서 매매, 가격 차이 실현
3. 이익창출 
<br>
시장의 효율성을 유지하는 데 중요한 역할을 합니다. <br>

- 효율적인 시장에서는 무차익 거래 기회가 희박하며, 이는 시장 참여자들이 가격 정보를 신속하게 반영하고 이익을 추구하려는 경쟁적인 환경을 의미합니다.
- 무차익 거래 기회가 발생할 수 있습니다. 이는 시장의 불완전성, 정보 비대칭, 거래 시스템의 지연 등 다양한 요인에 의해 발생할 수 있습니다. 
- 따라서 무차익 거래는 시장의 효율성을 유지하고 시장 참여자들의 경쟁을 촉진하는 중요한 요소로 간주됩니다.

```python
import numpy as np

S0 = 10
B0 = 10
print("오늘의 주식 채권 :",S0,B0)
#오늘의 주식, 채권 가격
print("")
S1 = np.array((20, 5))
B1 = np.array((11,11))
print("내일의 주식, 채권 가격 :", S1, B1)
print("")
#내일의 주식과 채권 가격의 페이오프

M0 = np.array((S0,B0))
print("시장가격 :", M0)
#시장가격 벡터
print("")
M1 = np.array((S1,B1)).T
print(M1)
print("")
#시장 페이오프 정렬

K = 14.5 
#옵션의 행사가

C1=np.maximum(S1-K,0) #최대값
#옵션의 불확실 페이오프 옵션을 행사하지 않으면 0
print("C1 : ",C1)
print("")
phi = np.linalg.solve(M1, C1) #numpy.linalg.solve() 함수는 선형 방정식의 해를 구하는 데 사용
#복제 포토플리오
print("phi:",phi)
print("")
print(np.allclose(C1, np.dot(M1, phi))) #numpy.allclose() 함수는 두 배열이 거의 같은지를 확인하는 함수
#복제 포토플리오의 페이오프가 옵션의 페이오프와 같은지 확인
print("")
C0 = np.dot(M0, phi) #numpy.dot() 함수는 두 배열의 내적(행렬 곱셈)을 계산
#복제포토플리오의 가격은 옵션의 무차익 거래와 가격과 같음
print(C0)
```

### 투자자의 선호도<br> ###
선호도가진 투자자가 경제모델 $M^2$ 에서 거래 가능한 두 자산에 투자한다고 전제<br>
투자자의 선호도는 미래의 페이오프에 대해 정의됨<br>
A < B(B 선호), A > B(A 선호), A ~ B(무차별)<br>
1. 완전성 - 투자자가 모든 페이오프에 대해 순위를 매길 수 있다. A > B, A < B, A ~ B
2. 이행성 - 미래의 페이오프 C를 가진 제 3의 포토플리오가 존재한다면 A > B & B > C 일 때 A > C이다.
3. 연속성 - A > B > C이면 B ~ &alpha;A + (1 - &alpha;)C가 성립하는 숫자 $alpha$ [0,1]이 존재한다.
4. 독립성 - A ~ B이면 &alpha;A + (1 - &alpha;)C ~ &alpha;B + (1 - &alpha;)C가 성립한다.<br> 
마찬가지로 A > B이면 &alpha;A + (1 - &alpha;)C > &alpha;A + (1 - &alpha;)C가 성립한다.
5. 지배성 - C1 = &alpha;1A + (1 - &alpha;1)이고 C2 = &alpha;2A + (1 - &alpha;2) 이면 A > C이고 C1 > C2 일 때 &alpha;1 > &alpha;2가 성립한다.<hr>
순서에 주목해야함

### 리스크 회피 ###
리스크 회피 측도 
- 절대 리스크 회피 : 투자자가 투자 포트폴리오에서 발생할 수 있는 모든 리스크를 완전히 제거하려는 전략<br>
- 애로우 프랫 측도 : 행동 경제학에서 사용되는 개념, 위험을 다룰 떄 사용, 어떤 자산에 투자할 때 그 자산의 변동성에 따라서 효용이 어떻게 변화하는지 알아보는데 사용<br>
<br>
금융이론과 모델에서는 리스크 회피와 중립이 나타날 것으로 가정(도박을 선호)<br>
$u(x)$ = $ln(x)$ - 리스크 회피, $u(x)$ = $x$ - 리스크 중립, $u(X)$ = $x^2$ - 리스크 선호<br>


- 예제 : 앞에 경제모델에 대해서 생각하면서 선호도가 있는 투자자가 미래의 페이오프 사이의 기대효용에 따라 의서결정 한다.<br>
u(x) = x<sup>1/2</sup> 라면, 포토플리오 D1의 페이오프보다 A1의 페이오프를 선호한다. 

```python

S0 = 10
B0 = 10
M0 = np.array((S0,B0))

def u(x) :
    return np.sqrt(x)
#리스크 회피 베르누이 함수

phi_A = np.array((0.75, 0.25))
phi_D = np.array((0.25, 0.75))
#서로 다른 가중치를 가진 포토플리오
print('')

print(np.dot(M0, phi_A) == np.dot(M0, phi_D)) 
#두 포토플리오를 만드는 비용은 같음
print("")
a1 = np.dot(M1, phi_A) #numpy.dot() 함수는 두 배열의 내적(행렬 곱셈)을 계산
print(a1)#한 포토플리오의 페이오프
print("")
d1 = np.dot(M1, phi_D)
print(d1)#다른 포토플리오의 페이오프
print("")
P = np.array((0.5, 0.5)) #확률 측도

def eut(x) :
    return np.dot(P,u(x))
#기대효용함수

print("a1:", eut(a1))
print("")
print("d1:",eut(d1))
#두 페이오프의 효용치
#A를 선호
```

이 상황에서 나올 수 있는 문제는 고정 비용으로 양의 가중치를 가진 최적의 포토플리오를 찾는 것<br>
결과 -> 위험자산 60%, 무위험 자산 40%를 배분

```python
from scipy.optimize import minimize
w = 10 #고정된 예산
cons = {'type':'eq', 'fun':lambda phi : np.dot(M0,phi) - w} #minimize 함수에 넣을 제한 조건
def eut_(phi) :
    x = np.dot(M1, phi) # 포토플리오에 대햇 정의한 기대효용함수
    return eut(x)

opt = minimize(lambda phi: -eut_(phi), x0 = phi_A, constraints=cons) # type - 등식의 제한조건, fun - 최소화정의 함수, x0 - 최적화를 시작할 초기 추정치, 
#eut(phi)를 최대화하기 위해 -eut(phi)를 최소화, x0 - 초기값, 예산제한조건
#최적화 문제를 해결하기 위한 도구, 주어진 함수를 최소화하는 최적화 알고리즘 실행하여 최적의 해를 찾음

print(opt)
print("---------------------")
print(eut_(opt['x'])) # 예산 w = 10에 대한 최적(최고) 기대효용, opt['x']에 가중치가 저장 
```

## 3.3 평균-분산 포토플리오 이론 ##

### 3.3.1 가정과 결과 ###
- 핵심가정 - 투자자들은 기대수익과 수익률의 분산에만 관심을 둔다. 
기대수익이 최대인 포토플리오의 분산이 반드시 최소인 것은 아니다.<br> 투자자의 선택에 따라 분산이 변함<br>
투자 포토플리오가 가져올 것으로 예상되는 수익의 1차/2차 모멘트로 에이전트의 선호도와 효용함수를 정의

1. 포토플리오 통계<br>
<br>
정적인 경제 모델 정의, A는 N개의 거래가능한 자산의집합, A0은 오늘 가격, A1은 1년 후 가격<br><- 수익률벡터<br>
모든 미래의 상태가 같은 확률을 가지는 경우 n번째 자산의 기대 수익률<br><br>
기대수익률벡터<br><br>포토플리오 벡터는 각 자산에 대해 가중치를 준 값으로 각각은 0이상, 합은 1<br>
<br>가중치 벡터와 포토플리오 기대 수익률 벡터의 곱 - 포토플리오 기대 수익률<br>
자산n과 자산m 이렇게 두 자산 간의 공분산 정의<br><br>
공분산 행렬<br><br>
포토플리오의 분산<br><br>
포토플리오의 변동성(표준편차)<br><br>

2. 샤프비율<br>
<br>
펀드나 다른 포토플리오 혹은 단일 자산에 대해 위험 조정 성능을 측정할 수 있는 수단.  투자의 수익 대비 리스크를 측정하는 데 사용.  포토플리오의 기대 수익률을 변동성으로 나눈 값<br>
무위험 이자율 r을 고려하면 포토플리오의 리스크 프리미엄, 초과 수익률 = 수익률 - 무위험이자율<br>
<br>
무위험이자율이 상대적으로 낮을 때는 차이 x<br>

1. 높은 샤프 비율: 높은 샤프 비율은 투자자에게 높은 수준의 초과 수익을 달성하는 데 대한 단위 변동성(리스크)를 얻었음을 나타냅니다.<br> 따라서 높은 샤프 비율은 좋은 투자 성과를 의미할 수 있습니다.<br>
2. 낮은 샤프 비율: 낮은 샤프 비율은 단위 변동성에 비해 상대적으로 낮은 초과 수익을 달성했음을 나타냅니다. 이는 투자의 리스크 조정된 수익이 낮다는 것을 의미하며, 투자 성과가 높지 않다는 것을 나타낼 수 있습니다.

```python
#3.3.2 수치예제

rS = S1/S0 - 1 #위험 자산의 수익률 벡터
rB = B1/B0 - 1 #안전 자산의 수익률 벡터

def mu(rX):
    return np.dot(P, rX) #기대 수익률 함수

print("rs :",mu(rS)) #기대 수익률
print("rb :",mu(rB)) #기대 수익률

rM = M1/M0 -1 #수익률 행렬
print("rM :",rM) 
print("기대수익률 벡터" ,mu(rM)) #기대 수익률 벡터
```

```python
#분산, 변동성, 공분산 행렬
def var(rX) :
    return((rX - mu(rX)) ** 2).mean()

print("Var(rS) :", var(rS))
print("var(rB) :", var(rB))

def sigma(rX) :
    return np.sqrt(var(rX))

print('변동성 rS :',sigma(rS))
print("변동성 rB : ", sigma(rB))

np.cov(rM.T, aweights=P, ddof=0)
```

```python
#모든 가중치가 같은 포토플리오의 기대 수익률, 분산,변동성

phi = np.array((0.5, 0.5))

def mu_phi(phi) :
    return np.dot(phi, mu(rM)) #포토플리오 기대 수익률

print(mu_phi(phi))

def var_phi(phi) :
    cv = np.cov(rM.T, aweights=P, ddof=0)
    return np.dot(phi, np.dot(cv,phi)) #포토플리오 분산

print(var_phi(phi))

def sigma_phi(phi) :
    return var_phi(phi) ** 0.5 #포토플리오의 변동성

print(sigma_phi(phi))
```

### 투자 기회 집합 ###
- 포토플리오 가중치에 대한 몬테카를로 시뮬레이션(확률적인 방법을 사용하여 불확실성이나 랜덤성을 포함하는 문제를 해결하는 데 사용되는 통계적 기법)을 기반으로 다양한 변동성-수익률 공간의 투자 기회 집합을 시각화 가능하다.

- 최소 변동성 및 최대 샤프 비율<br>
위험 자산 T는 기대 수익률이 마이너스지만 최대 샤프 비율 포토플리오에서는 비중이 높다.<br>
이는 다변화 효과로 인해 포토플리오의 예상 수익률보다 포토플리오 리스크가 더 낮아졌기 때문이다.

```python
cons = {'type': 'eq', 'fun': lambda phi: np.sum(phi) - 1}
bnds = ((0, 1), (0, 1))
min_var = minimize(sigma_phi, (0.5, 0.5),
                   constraints=cons, bounds=bnds)

print(min_var)
print("-------------------------")
def sharpe(phi):
    return mu_phi(phi) / sigma_phi(phi) #샤프비율함수를 정의

max_sharpe = minimize(lambda phi: -sharpe(phi), (0.5, 0.5),
               constraints=cons, bounds=bnds) 
#음의 값을 최소화함으로써 샤프 비율을 최대화

print(max_sharpe)
print("-------------------------")

plt.figure(figsize=(10, 6))
plt.plot(mcs[:, 0], mcs[:, 1], 'ro', ms=5)
plt.plot(sigma_phi(min_var['x']), mu_phi(min_var['x']),
         '^', ms=12.5, label='minimum volatility')
plt.plot(sigma_phi(max_sharpe['x']), mu_phi(max_sharpe['x']),
         'v', ms=12.5, label='maximum Sharpe ratio')
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.legend();
```

- 효율적 투자 경계선<br>
효율적 포토플리오는 예상되는 리스크를 고려할 때 최대의 기대수익을 가지는 포토플리오<br>
최소 위험 포토플리오보다 낮은 기대 수익률을 보이는 모든 포토플리오는 비효율적<br>
아래 코드 리스크-수익률 공간에서 효율적인 포토플리오 도출, 곡선 시각화<br>
효율적인 모든 포토플리ㅗ 집합 - 효율적 투자 경계선, 투자자 이 안에서 선택

```python
cons = [{'type': 'eq', 'fun': lambda phi: np.sum(phi) - 1},
       {'type': 'eq', 'fun': lambda phi: mu_phi(phi) - target}]
#제안 조건으로 기대 수익률 목푯값 설정
bnds = ((0, 1), (0, 1))
targets = np.linspace(mu_phi(min_var['x']), 0.16) #목표 기대 수익률 집합 생성

frontier = []
for target in targets:
    phi_eff = minimize(sigma_phi, (0.5, 0.5),
                       constraints=cons, bounds=bnds)['x'] #주어진 기대 수익률에 대해 가장 변동성이 작은 포토플리오 선택
    frontier.append((sigma_phi(phi_eff), mu_phi(phi_eff)))
frontier = np.array(frontier)

plt.figure(figsize=(10, 6))
plt.plot(frontier[:, 0], frontier[:, 1], 'mo', ms=5,
         label='efficient frontier')
plt.plot(sigma_phi(min_var['x']), mu_phi(min_var['x']),
         '^', ms=12.5, label='minimum volatility')
plt.plot(sigma_phi(max_sharpe['x']), mu_phi(max_sharpe['x']),
         'v', ms=12.5, label='maximum Sharpe ratio')
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.legend();
```

```python
phi_M = np.array((0.8, 0.2))
mu_M = mu_phi(phi_M)
print('mu_M :',mu_M)

sigma_M = sigma_phi(phi_M)
print('sigma_M :', sigma_M)

r = 0.0025

plt.figure(figsize=(10, 6))
plt.plot(frontier[:, 0], frontier[:, 1], 'm.', ms=5,
         label='efficient frontier')
plt.plot(0, r, 'o', ms=9, label='risk-less asset')
plt.plot(sigma_M, mu_M, '^', ms=9, label='market portfolio')
plt.plot((0, 0.6), (r, r + ((mu_M - r) / sigma_M) * 0.6),
         'r', label='capital market line', lw=2.0)
plt.annotate('$(0, \\bar{r})$', (0, r), (-0.015, r + 0.01))
plt.annotate('$(\sigma_M, \mu_M)$', (sigma_M, mu_M),
             (sigma_M - 0.025, mu_M + 0.01))
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.legend();
```

- 최적 포토플리오 <br>
<Br>

- 이차효용함수<br>
자본자산 가격결정 모형 & 포토플리오 이론 - 투자자가 단일 기간동안 수익률, 리스크에만 관심있다 가정 - 2차 베르누이 효용함수의 형태일 때만 적용<br>
평균-분산 포플에서만 사용<br>
이차효용함수나 정규분포 자산수익률 분포 가정 - 좋은 방법 X

투자자의 포토플리오 조합? - 효용함수로 파악 가능 <br> b = 1 고정

```python
def U(p):
    mu, sigma = p
    return mu - 1 / 2 * (sigma ** 2 + mu ** 2) 
#리스크-수익률 공간상의 효용함수

cons = {'type': 'eq',
        'fun': lambda p: p[0] - (r + (mu_M - r) / sigma_M * p[1])}
#포토플리오가 자본시장선에 있을 조건 

opt = minimize(lambda p: -U(p), (0.1, 0.3), constraints=cons)
opt
```

### 무차별 곡선 ###
시각적 분석을 통해 최적의 의사결절 설명 가능<br>
리스크-수익률 공간에서 무차별 곡선 그리기 가능<br>
최적의 포토플리오 - 자본시장선과 접하는 위치

심볼연산을 통해 리스크-수익률 공간 효용함수, 고정된 효용 수준 v와 고정된 파라미터 &beta;에 대한 u와 &alpha; 의 함수 관계로 변환<br>
두 개의 모차별 곡선을 보여주면서 모든 조합에 대해 무차별 곡선을 그림<br>
투자ㅈ는 하나의 무차별 곡선 내의 포토플리오에 대해 차별 X

```python
from sympy import *
init_printing(use_unicode=False, use_latex=False)

mu, sigma, b, v = symbols('mu sigma b v')
sol = solve('mu - b / 2 * (sigma ** 2 + mu ** 2) - v', mu)
#u에 대한 효용함수 구하기
sol
```

```python
u1 = sol[0].subs({'b': 1, 'v': 0.1})
#수치값 대입
u1
```

```python
u2 = sol[0].subs({'b': 1, 'v': 0.125})
#수치값 대입
u2
```

```python
f1 = lambdify(sigma, u1)
f2 = lambdify(sigma, u2)
#호출가능함수 생성
sigma_ = np.linspace(0.0, 0.5)
#함수를 호출할 알파값지정
u1_ = f1(sigma_)
u2_ = f2(sigma_)
#두 가지 다른 효용 수준에 대해 함수 호출
plt.figure(figsize=(10, 6))
plt.plot(sigma_, u1_, label='$v=0.1$')
plt.plot(sigma_, u2_, '--', label='$v=0.125$')
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.legend();
```

무차별 곡선을 자본시장선과 결합하여 투자자의 초적 포토플리오 파악<br>
앞서 수치 최적화 한 결과를 통해 무차별 곡선(최적의 포플)이 자본시장선에 접하는 위치 시각화<br>
실제로 시장 포토플리오와 무위험 자산을 혼합하여 선택을 시사

```python
u = sol[0].subs({'b': 1, 'v': -opt['fun']})
#최적효용함수에 대한 무차별 곡선 정의
u
```

```python
f = lambdify(sigma, u)
u_ = f(sigma_)
#무차별 곡선을 그리기 위한 수치값 계산

plt.figure(figsize=(10, 6))
plt.plot(0, r, 'o', ms=9, label='risk-less asset')
plt.plot(sigma_M, mu_M, '^', ms=9, label='market portfolio')
plt.plot(opt['x'][1], opt['x'][0], 'v', ms=9, label='optimal portfolio')
plt.plot((0, 0.5), (r, r + (mu_M - r) / sigma_M * 0.5),
         label='capital market line', lw=2.0)
plt.plot(sigma_, u_, '--', label='$v={}$'.format(-round(opt['fun'], 3)))
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.legend();
```

 <br>

두 개의 위험 자산 뿐만 아니라 다음과 같은 미래 페이오프를 가지는 세번째 자산 V<br>
<br>
S1, T1만으로 R3 기저벡터를 이룰 수 없다. 최소자승 회귀를 사용하여 페이오프 V1에 대한 유사값 사용 가능

```python
#최소자승회귀를 통해 구현
M1
print("m1 : ",M1)
M0
print("M0 :", M0)
V1 = np.array((12, 15, 7))
reg = np.linalg.lstsq(M1, V1, rcond=-1)[0] #선형대수모듈 사용하여 최소제곱회귀 수행
#최소 회귀 인숫값을 요인적재량으로 해석
print(reg)
np.dot(M1, reg)
print(np.dot(M1, reg))
np.dot(M1, reg) - V1
print(np.dot(M1, reg) - V1)
V0 = np.dot(M0, reg)
print('V0 :', V0)
#요인적제량으로 자산 V에 대한 무차익거래 가격 추정 가능
```

두개의 요인 - 완벽하게 설명 불가<br>이 경제 모델에 세번째 리스크 모델 요인 U 추가, U0 = 10<br>
<br>
3개의 리스크 요인으로 페이오프 V1을 완벽하게 복제 가능

```python
U0 = 10
U1 = np.array((12, 5, 11))
M0_ = np.array((S0, T0, U0))
#증가한 시장가격 벡터
M1_ = np.concatenate((M1.T, np.array([U1,]))).T
#풀랭크로 증가한 시장 페이오프 행렬
print('M1 :', M1_)
np.linalg.matrix_rank(M1_)
reg = np.linalg.lstsq(M1_, V1, rcond=-1)[0]
print('reg :',reg)
print(np.allclose(np.dot(M1_, reg), V1))
#V1의 정확한 복제, 잔차 0
V0_ = np.dot(M0_, reg)
print('V0_ :',V0_)
#위험 자산 V0에 대한 유일한 무차익거래 가격
```

충분한 수의 리스크 요인을 사용하여 무차익거래 가격을 유도하였다는 점에서 3.1 예제와 유사<br>
차익거래 가격결정 이론 = 잔차가 있어도 된다. 하지만 충분한 개수의 리스크 요인이 있으면 잔차가 0되는 복제 가능<hr>
